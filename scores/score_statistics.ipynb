```python
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns
import os
sns.set(color_codes=True)
```

```python
# draw score distribution
def draw_dist(root, score_file):
    with open(os.path.join(root, score_file)) as f:
        lines = f.readlines()
    scores = [float(i) for i in lines]
    sns.distplot(scores, norm_hist=False)
```

```python
# count samples in each bucket
def count_samples(root, bucket_file):
    with open(os.path.join(root, bucket_file)) as f:
        lines = f.readlines()
    scores = [int(i) for i in lines]
    nb = len(set(scores))
    counts = [0] * nb
    for s in scores:
        counts[s] += 1
    print(bucket_file[:-3] + "\n")
    print("The sample counts are " + ",".join([str(c) for c in counts]) + " in " + str(nb) + " buckets respectively.")
    print("The distribution of scores are " + ",".join(["{:.2}".format(float(c)/sum(counts)) for c in counts]))
```
